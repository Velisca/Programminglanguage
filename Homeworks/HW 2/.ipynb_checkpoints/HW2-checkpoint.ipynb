In this analysis, I'm going to answer these 3 questions below:
1. In average, how long does it take one person to get promoted?
2. Does being promoted means having higher salary?
3. Does one's background make some difference in the rate of being promoted?

train.csv is the first dataset showing the rate of promotion affected by several factors like education, length of services, ratings, achievements, age and trainings. 

```python
import pandas as pd
data = pd.read_csv('train.csv')
data.head()
```

```python
oldcolumn_one = 'years_spent'
newcolumn_one = 'years'
oldcolumn_two = 'promotion'
newcolumn_two = 'promotion_rate'
oldcolumn_three = 'division'
newcolumn_three = 'department'
data.rename(columns={oldcolumn_one: newcolumn_one}, inplace=True)
data.rename(columns={oldcolumn_two: newcolumn_two}, inplace=True)
data.rename(columns={oldcolumn_three: newcolumn_three}, inplace=True)
```

```python
change_datacolumns = 'train.csv'
data.to_csv(change_datacolumns, index=False)
updated_data = pd.read_csv(change_datacolumns)
```

HR_Dataset.csv is the second dataset showing the salary of employees from several departments after keeping in check of several factors like projects handled, working hours, promotion and evaluation score in the past 5 years.

I'm going to combine both dataset

```python
file1 = 'train.csv'
chosen_columns1 = ['department', 'education', 'no_of_trainings', 'years', 'avg_training_score', 'promotion_rate']
df1 = pd.read_csv(file1, usecols=chosen_columns1)

file2 = 'HR_Dataset.csv'
chosen_columns2 = ['number_project', 'years_spent', 'promotion', 'division', 'salary'] 
df2 = pd.read_csv(file2, usecols=chosen_columns2)

combined_df = pd.concat([df1, df2], axis=1)
combined_df.head()
```

```python
print(combined_df)
```

```python
columns_to_convert = ['number_project', 'years_spent', 'promotion']

for column in columns_to_convert:
    combined_df[column] = combined_df[column].fillna(0).astype(int)

print(combined_df)
```

```python
print(combined_df.columns)
```

Like what is seen above, there are two years_spent columns and two promotion columns.
I'm going to combine both columns

```python
combined_df.head()
```

```python
column_data_types = combined_df.dtypes

print(column_data_types)
```

```python
combined_df['experience'] = combined_df['years_spent'] + combined_df['years']
combined_df['successful_promotion'] = combined_df['promotion'] + combined_df['promotion_rate']
combined_df['departments'] = combined_df['department'] + combined_df['division']
```

```python
combined_df.head()
```

```python
print(combined_df.columns)
```

Next, I'm going to only exclude the years_spent and promotion to be able to answer the first question better.  
1. In average, how long does it take one people to get promoted?

```python
new_df = combined_df[['experience', 'successful_promotion']]
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
new_df.head()
```

```python
print(new_df[['experience', 'successful_promotion']])
```

```python
promoted_employees = new_df[new_df['successful_promotion'] == 1]

print(promoted_employees)
```

```python
twice_promoted_employees = new_df[new_df['successful_promotion'] == 2]

print(twice_promoted_employees)
```

```python
missing_data = new_df.isnull() 
missing_counts = missing_data.sum()
print(missing_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(new_df, x='successful_promotion', y='experience')
plt.xlabel('Promotion Status')
plt.ylabel('Years Spent')
plt.title('Distribution of Promotion Status')
plt.xticks([0, 1,2], ['Not Promoted', 'Promoted','Promoted Twice'])
plt.show()
```

Here, we can see that even if longer years of service, promotion is not by default. 

Now, I'll move on to the next question      
2. Does being promoted means having higher salary?

First, I combining the columns I needed into a new dataframe to be able to see the results better

```python
new2_df = combined_df[['successful_promotion', 'salary']]
new2_df.head(15)
```

Next, I'm checking if there are any missing values and deleting them if there is any

```python
missing_data2 = new2_df.isnull() 
missing_counts2 = missing_data2.sum()
print(missing_counts2)
```

```python
new2_df_cleaned = new2_df.dropna()
```

```python
missing_data2 = new2_df_cleaned.isnull() 
missing_counts2 = missing_data2.sum()
print(missing_counts2)
```

After cleaning the data, it's time to depict it in a graph

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(new2_df, x='salary', hue='successful_promotion', order=['low', 'medium', 'high'])
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Distribution of Salary by Promotion Status')
plt.legend(title='Promotion', labels=['Not Promoted', 'Promoted', 'Promoted Twice'])
plt.show()
```

Here, we can see that despite being promoted, the difference in salary is not that significant. Most people still receive low and medium salary levels despite having higher status.

Heading to the last question      
3. Does one's background make some difference in the rate of being promoted?

Here, by means of background, I mean education, number of projects, average training years, and training scores

```python
new3_df = combined_df[['education', 'no_of_trainings', 'avg_training_score', 'number_project', 'successful_promotion']]
new3_df.head(15)
```

Checking and deleting missing data

```python
missing_data3 = new3_df.isnull() 
missing_counts3 = missing_data3.sum()
print(missing_counts3)
```

```python
new3_df_cleaned = new3_df.dropna()
```

```python
missing_data3 = new3_df_cleaned.isnull() 
missing_counts3 = missing_data3.sum()
print(missing_counts3)
```

```python
new3_df.head(10)
```

Now that it's done, I'm going to illustrate it in the chart to show the comparison

```python
plt.figure(figsize=(20, 12))

# Box plot for Education
plt.subplot(221)
sns.boxplot(new3_df, x='education', y='successful_promotion')
plt.title('Education vs. Promotion')

# Box plot for Number of Projects
plt.subplot(222)
sns.boxplot(new3_df, x='number_project', y='successful_promotion')
plt.title('Number of Projects vs. Promotion')

# Box plot for Average Training Score
plt.subplot(223)
sns.boxplot(new3_df, x='avg_training_score', y='successful_promotion')
plt.title('Average Training Score vs. Promotion')

# Box plot for Training Years
plt.subplot(224)
sns.boxplot(new3_df, x='no_of_trainings', y='successful_promotion')
plt.title('Training Years vs. Promotion')

plt.tight_layout()
plt.show()
```

Looking at these 4 comparison charts, we can make sum up the conclusions into some points below:   
- People use to believe that with a Master degree, the chances of being promoted will be higher. But here, we can see that it's not certain.  
- In the next chart, we can also see that the number of project handled by a person don't always guarantee promotion.  
- If one score badly during training period, it doesn't mean that the chance of promotion will be 0.  
- Receiving more training doesn't mean that one will be guaranteed to be promoted.
