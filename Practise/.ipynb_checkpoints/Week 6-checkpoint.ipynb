```python
import urllib.request as req 
import json
import csv
import pandas as pd
import re
import bs4
```

```python
url = "https://www.lyrics.com/lyric/35002717/Anne-Marie"
with req.urlopen(url) as response:
    data = response.read().decode("utf-8")
print(data)
```

```python
home = bs4.BeautifulSoup(data, "html.parser")
```

```python
title_element = home.find("element_selector")

# Check if the title_element is not None and extract the title
if title_element:
    title = title_element.text
    print(title)
else:
    print("Title not found")
```
